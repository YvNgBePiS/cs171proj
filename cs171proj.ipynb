```python
from fancyimpute import KNN
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline

cols = ['Application mode', 'Course', 'Previous qualification', "Mother's qualification", "Father's qualification", 'Tuition fees up to date', 
        "Mother's occupation", "Father's occupation", 'Gender', 'Scholarship holder', 'Age at enrollment', 'Curricular units 1st sem (approved)', 
        'Curricular units 2nd sem (approved)', 'Target']

df = new_df[cols]

df = df[df['Target'] != 'Enrolled']

df = df.replace({'Target': {'Dropout': 0, 'Graduate': 1}})

df['Tuition fees up to date'] = KNN(k=3).fit_transform(df[['Tuition fees up to date']])
df['Scholarship holder'] = KNN(k=3).fit_transform(df[['Scholarship holder']])

impute_cat_cols = ['Application mode', 'Course', 'Previous qualification']

for col in impute_cat_cols:
    features = df.dropna().drop(col, axis=1)
    target = df.dropna()[col]

    imputer = make_pipeline(SimpleImputer(strategy='most_frequent'), DecisionTreeClassifier())
    imputer.fit(features, target)

    df[col] = df.apply(lambda row: row[col] if not pd.isna(row[col]) else imputer.predict([row.drop(col)])[0], axis=1)

impute_unit_cols = ['Curricular units 1st sem (approved)', 'Curricular units 2nd sem (approved)']

for col in impute_unit_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)

cols = ['Tuition fees up to date', 'Gender', 'Scholarship holder', 'Target']
df[cols] = df[cols].astype('int32')
```
